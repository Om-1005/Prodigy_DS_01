```python
## dropping unecessary columns
ds.drop(['Series Name','Country Code'],axis=1)
```

```python
#Histogram of total population
total_population_data = ds[ds['Series Code']=='SP.POP.TOTL'].copy()
numerical_columns = total_population_data.select_dtypes(include='number').columns
total_population_data['Row Total'] = total_population_data[numerical_columns].sum(axis=1)
plt.hist(total_population_data['Row Total'])
plt.xlabel('total population')
plt.ylabel('Frequency')
plt.title('distribution of Total Population')
plt.show()
```

```python
#Extraction of top 10 countries with respect to population
total_population_sort = total_population_data.sort_values(by="Row Total",ascending= False)
top_countries = total_population_sort.head(10)
print("Top 10 countries with row totals of the population total column\n")
print(top_countries[["Country Name","Row Total"]])

```

```python
# Bar Plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="Row Total",y="Country Name",data=top_countries,palette="plasma")
plt.title("Population of top 10 countries",fontsize = 10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country ",fontsize=10)
plt.show()
```

```python
#Scatter plot
#Top 10 countries of population
plt.scatter(x="Row Total",y="Country Name",data= top_countries,color="red")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.title("Scatter Plot")

```

```python
#Histogram of Total Male Population
male_population = ds[ds['Series Code']=='SP.POP.TOTL.MA.IN'].copy()
num_values = male_population.select_dtypes(include="number").columns
male_population["Row Total"]= male_population[num_values].sum(axis=1)
plt.hist(male_population["Row Total"])
plt.title('Distribution male population')
plt.xlabel("total number of male population")
plt.ylabel("Frequency")
```

```python
# Extraction of top 10 countries with respect to male population
male_population_sorted = male_population.sort_values(by="Row Total",ascending = False)
male_population_top = male_population_sorted.head(10)
print("Total male population of top 10 countries \n")
print(male_population_top[['Country Name','Row Total']])
```

```python
#Bar plot
plt.figure(figsize=(15,6))
sns.barplot(x="Row Total",y="Country Name",data=male_population_top,palette="plasma")
plt.title("Bar plot of male Population of top 10 countries")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()
```

```python
#Scatter Plot
plt.scatter(x="Row Total",y="Country Name",data=male_population_top,color="green")
plt.title("Scatter plot of Male Population")
plt.xlabel("total male population")
plt.ylabel("Country")
plt.show()
```

```python
#Histogram Population of Female
female_population = ds[ds["Series Code"]=="SP.POP.TOTL.FE.IN"].copy()
numm_values= female_population.select_dtypes(include="number").columns
female_population['row total'] = female_population[numm_values].sum(axis=1)
plt.hist(female_population['row total'])
plt.title("Distribution of Female Population")
plt.xlabel("Total Population")
plt.ylabel("Frequency")
plt.show()
```

```python
# Extracting the top 10 countries with respect to Female population
female_population_sorted = female_population.sort_values(by="row total",ascending= False)
female_top = female_population_sorted.head(10)
print("Top 10 countries with respect to femaale population\n")
print(female_top[["Country Name","row total"]])
```

```python
# Bar plot
plt.figure(figsize=(15,6))
sns.barplot(x="row total",y="Country Name",data=female_top,palette='plasma')
print("top ten countries female population total \n")
plt.title("Top Ten Countries of Female Population",fontsize=10)
plt.xlabel("Total female Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
## scatter plot
plt.scatter(x="row total", y="Country Name", data=female_top,color='red')
plt.title("Scatter Plot",fontsize=10)
plt.xlabel("Total Female Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
